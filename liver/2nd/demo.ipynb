```python
df["gender"] = df['gender'].replace({
  "Male":0,
  "Female":1
})
df["cancer_stage"] = df["cancer_stage"].replace({
    "Stage I": 1,
    "Stage II": 2,
    "Stage III": 3,
    "Stage IV": 4
})
df["family_history"] = df['family_history'].replace({
  "No":0,
  "Yes":1
})
df["smoking_status"] = df["smoking_status"].replace({
  'Never Smoked':0,
  'Passive Smoker':1, 
  'Former Smoker': 2,
  'Current Smoker': 3
})
df["treatment_type"] = df["treatment_type"].replace({
  'Combined':0, 
  'Radiation':1, 
  'Surgery':2, 
  'Chemotherapy':3
})
```

```python
df = df[['age', 'gender', 'cancer_stage',
        'family_history', 'smoking_status',
       'bmi', 'cholesterol_level', 'hypertension', 'asthma', 'cirrhosis',
       'other_cancer', 'treatment_type', 'survived']]
```

```python
plt.figure(figsize=(20,8))
sns.heatmap(df.corr(),annot=True,linewidths=.5)
```

```python
a = df.bmi.unique()
pd.Series(a).sort_values()
```

```python
IQR = df.bmi.quantile(0.75) - df.bmi.quantile(0.25)

upper_bridge = df.bmi.mean() + 3* df.bmi.std()
lower_bridge = df.bmi.mean() - 3* df.bmi.std()

print(f"upper_bridge : {upper_bridge}\nlower_bridge : {lower_bridge}")
df.bmi = np.where(
    df.bmi >= upper_bridge, upper_bridge, df.bmi
)
```

```python
IQR = df.bmi.quantile(0.75) - df.bmi.quantile(0.25)
upper_bridge = df.bmi.quantile(0.75) + (IQR*3)
lower_bridge = df.bmi.quantile(0.25) - (IQR*3)

print(f"upper_bridge : {upper_bridge}\nlower_bridge : {lower_bridge}")

df.bmi = np.where(df.bmi >= upper_bridge, upper_bridge, df.bmi)
```

```python
plt.figure(figsize=(20,8))
sns.heatmap(df.corr(),annot=True,linewidths=.5)
```

```python
df = df.drop(["bmi"],axis=1)
```

```python
plt.figure(figsize=(20,8))
sns.heatmap(df.corr(),annot=True,linewidths=.5)
```

```python
from imblearn.combine import SMOTETomek
X = df.iloc[:,:-1]
Y = df.iloc[:,-1]
```

```python
smo_tek = SMOTETomek()
X_smotek, y_smotek = smo_tek.fit_resample(X, Y)
```

```python
from collections import Counter
Counter(Y)
Counter(y_smotek)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_smotek, y_smotek, test_size=0.25, random_state=33)
```

```python
X_train.shape
```

```python
y_train
```

```python
X_train
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(X_train,y_train)
y_pred_knn = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred_knn))
print(classification_report(y_test, y_pred_knn))
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
y_pred_knn = knn.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred_knn))
print(classification_report(y_test, y_pred_knn))
```

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)
forest.fit(X_train, y_train)

y_pred_forest = forest.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred_forest))
print(classification_report(y_test, y_pred_forest))
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train, y_train)
y_pred_nb = nb.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred_nb))
print(classification_report(y_test, y_pred_nb))
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(max_depth=100)
tree.fit(X_train, y_train)
y_pred_tree = tree.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred_tree))
print(classification_report(y_test, y_pred_tree))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, max_depth=1, random_state=42)
gb.fit(X_train, y_train)
y_pred_gb = gb.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred_gb))
print(classification_report(y_test, y_pred_gb))
```

```python
from sklearn.neural_network import MLPClassifier
nl = MLPClassifier(max_iter=100)
nl.fit(X_train, y_train)
y_pred_nl = nl.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred_nl))
print(classification_report(y_test, y_pred_nl))
```
