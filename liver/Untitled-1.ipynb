```python
data.nunique()
```

```python
IQR = data.Alamine_Aminotransferase.quantile(0.75) - data.Alamine_Aminotransferase.quantile(0.25)

upper_bridge = data["Alamine_Aminotransferase"].quantile(0.75) + (IQR*3)
lower_bridge = data["Alamine_Aminotransferase"].quantile(0.25) - (IQR*3)

print(f"upper_bridge : {upper_bridge}\nlower_bridge : {lower_bridge}")
data["Alamine_Aminotransferase"] = np.where(
    data["Alamine_Aminotransferase"] > upper_bridge, upper_bridge, data["Alamine_Aminotransferase"]
)
```

```python
data.Total_Protiens.unique()
```

```python
IQR = data.Total_Protiens.quantile(0.75) - data.Total_Protiens.quantile(0.25)

upper_bridge = data["Total_Protiens"].mean() + 3* data["Total_Protiens"].std()
lower_bridge = data["Total_Protiens"].mean() - 3* data["Total_Protiens"].std()

print(f"upper_bridge : {upper_bridge}\nlower_bridge : {lower_bridge}")
data["Total_Protiens"] = np.where(
    data["Total_Protiens"] >= upper_bridge, upper_bridge, data["Total_Protiens"]
)
```

```python
IQR = data.Albumin_and_Globulin_Ratio.quantile(0.75) - data.Albumin_and_Globulin_Ratio.quantile(0.25)

upper_bridge = data["Albumin_and_Globulin_Ratio"].quantile(0.75) + (IQR*3)
lower_bridge = data["Albumin_and_Globulin_Ratio"].quantile(0.25) - (IQR*3)

print(f"upper_bridge : {upper_bridge}\nlower_bridge : {lower_bridge}")
data["Albumin_and_Globulin_Ratio"] = np.where(
    data["Albumin_and_Globulin_Ratio"] >= upper_bridge, upper_bridge, data["Albumin_and_Globulin_Ratio"]
)
```

```python
data.Albumin.unique()
```

```python
IQR = data.Albumin.quantile(0.75) - data.Albumin.quantile(0.25)

upper_bridge = data["Albumin"].mean() + 3* data["Albumin"].std()
lower_bridge = data["Albumin"].mean() - 3* data["Albumin"].std()

print(f"upper_bridge : {upper_bridge}\nlower_bridge : {lower_bridge}")

data["Albumin"] = np.where(
    data["Albumin"] >= upper_bridge, upper_bridge, data["Albumin"]
)
```

```python
from imblearn.combine import SMOTETomek
X = data.iloc[:,:-1]
Y = data.iloc[:,-1]
```

```python
smo_tek = SMOTETomek()
X_smotek, y_smotek = smo_tek.fit_resample(X, Y)
```

```python
from collections import Counter
Counter(Y)
Counter(y_smotek)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_smotek, y_smotek, test_size=0.25, random_state=33)
```

```python
X_train.shape
```

```python
y_train
```

```python
X_train
```

```python
test = X_train.Albumin_and_Globulin_Ratio.unique()
pd.Series(test).sort_values()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
model = LogisticRegression(max_iter=1000)
model.fit(X_train,y_train)
y_pred_knn = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred_knn))
print(classification_report(y_test, y_pred_knn))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
y_pred_knn = knn.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred_knn))
print(classification_report(y_test, y_pred_knn))
```

```python
from sklearn.svm import SVC

svm = SVC(kernel='linear')
svm.fit(X_train, y_train)
y_pred_svm = svm.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred_svm))
print(classification_report(y_test, y_pred_svm))
```

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)
forest.fit(X_train, y_train)

y_pred_forest = forest.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred_forest))
print(classification_report(y_test, y_pred_forest))
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train, y_train)
y_pred_nb = nb.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred_nb))
print(classification_report(y_test, y_pred_nb))
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(max_depth=5)
tree.fit(X_train, y_train)
y_pred_tree = tree.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred_tree))
print(classification_report(y_test, y_pred_tree))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, max_depth=1, random_state=42)
gb.fit(X_train, y_train)
y_pred_gb = gb.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred_gb))
print(classification_report(y_test, y_pred_gb))
```

```python
import streamlit as st
import pickle
import numpy as np


@st.cache_resource
def load_model():
    with open("model.pkl", "rb") as file:
        return pickle.load(file)

model = load_model()

st.header("Patient Data & Prediction")

age = st.slider("Age", 0, 100, 50)
gender = st.radio("Gender", ["Male", "Female"])
direct_bilirubin = st.slider("Direct Bilirubin", 0.0, 20.0, 10.0)
alkaline_phosphotase = st.slider("Alkaline Phosphotase", 0, 2000, 1000)
alamine_aminotransferase = st.slider("Alamine Aminotransferase", 0.0, 173.0, 100.0)
aspartate_aminotransferase = st.slider("Aspartate Aminotransferase", 0, 5000, 2500)
total_proteins = st.slider("Total Proteins", 0.0, 10.0, 5.0)
albumin = st.slider("Albumin", 0.0, 6.0, 3.0)
albumin_and_globulin_ratio = st.slider("Albumin and Globulin Ratio", 0.0, 3.0, 1.5)

gender_numeric = 1 if gender == "Male" else 0

input_data = np.array([
    age, gender_numeric, direct_bilirubin, alkaline_phosphotase,
    alamine_aminotransferase, aspartate_aminotransferase,
    total_proteins, albumin, albumin_and_globulin_ratio
]).reshape(1, -1)

st.write("### Collected Patient Data:")
st.json({
    "Age": age,
    "Gender": gender,
    "Direct Bilirubin": direct_bilirubin,
    "Alkaline Phosphotase": alkaline_phosphotase,
    "Alamine Aminotransferase": alamine_aminotransferase,
    "Aspartate Aminotransferase": aspartate_aminotransferase,
    "Total Proteins": total_proteins,
    "Albumin": albumin,
    "Albumin and Globulin Ratio": albumin_and_globulin_ratio
})


if st.button("Predict"):
    prediction = model.predict(input_data)[0]  

    if prediction == 1:
        st.error(" High Risk: The model predicts a potential liver disease.")
    else:
        st.success(" Low Risk: The model predicts a healthy liver condition.")
```
